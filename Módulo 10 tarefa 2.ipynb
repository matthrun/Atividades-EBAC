````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import sklearn

df = pd.read_csv('previsao_de_renda.csv')
print(df.head())
````

````python

# Criando o gráfico de ponto para posse_de_veiculo
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x='posse_de_veiculo', y='renda', data=df, dodge=True, errorbar=('ci', 90))

# Adicionando títulos e rótulos
plt.title('Renda por Posse de Veículo')
plt.xlabel('Posse de Veículo')
plt.ylabel('Renda Média')
plt.grid(True)
plt.show()
````

````python
# Criando o gráfico de ponto para posse_de_imovel
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x='posse_de_imovel', y='renda', data=df, dodge=True, errorbar=('ci', 90))

# Adicionando títulos e rótulos
plt.title('Renda por Posse de Imóvel')
plt.xlabel('Posse de Imóvel')
plt.ylabel('Renda Média')
plt.grid(True)
plt.show()
````
