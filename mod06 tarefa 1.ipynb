####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_bar(data, column, title):
    sns.countplot(x=column, data=data)
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel("Frequência")
    plt.show()

plot_bar(df, 'APGAR1', 'Frequência de APGAR1')
plot_bar(df, 'APGAR5', 'Frequência de APGAR5')
plot_bar(df, 'SEXO', 'Distribuição por Sexo')
plot_bar(df, 'GESTACAO', 'Distribuição por Gestação')
plot_bar(df, 'ESTCIVMAE', 'Estado Civil da Mãe')
```
