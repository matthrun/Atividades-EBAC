#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. #### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
print("Valores únicos em posse_de_veiculo:", df_sem_na.posse_de_veiculo.unique())
print("Valores únicos em mau:", df_sem_na.mau.unique())

# Mapeamento
df_sem_na.posse_de_veiculo = df_sem_na.posse_de_veiculo.map({'Y': 1, 'N': 0})
df_sem_na.mau = df_sem_na.mau.map({'True': 1, 'False': 0})

# NaN por 0 
df_sem_na.posse_de_veiculo.fillna(0, inplace=True)
df_sem_na.mau.fillna(0, inplace=True)

print(df_sem_na)
````

````python

# Tabela cruzada
tabela_cruzada = pd.crosstab(
    index=df_sem_na['possui_email'], 
    columns=df_sem_na['posse_de_veiculo'], 
    values=df_sem_na['mau'], 
    aggfunc='mean'
)

# Total de mau = 1 para cada combinação
tabela_cruzada_total = pd.crosstab(
    index=df_sem_na['possui_email'], 
    columns=df_sem_na['posse_de_veiculo'], 
    values=df_sem_na['mau'], 
    aggfunc='count'
)

tabela_cruzada = tabela_cruzada.fillna(0)
tabela_cruzada['total'] = tabela_cruzada_total.sum(axis=1)



tabela_cruzada['probabilidade_mau_1'] = tabela_cruzada[1] / tabela_cruzada['total']
print(tabela_cruzada)
````
