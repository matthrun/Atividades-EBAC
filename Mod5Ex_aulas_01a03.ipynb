```python
# 3) seu código aqui
colunas_interesse = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']
df = df[colunas_interesse]

contagem_missing = df.isna().sum()

print(df)
print(f'Contagem de valores missing:\n{contagem_missing}')
```

```python
# 4) Removendo registros onde Apgar5 não está preenchido

df_filtrado = df.dropna(subset=['APGAR5'])

numero_linhas = df_filtrado.shape[0]
contagem_missing = df_filtrado.isna().sum()

print(f'Número de linhas após remoção de registros com Apgar5 não preenchido: {numero_linhas}')
print(f'Contagem de valores missing após filtragem:\n{contagem_missing}')
```

```python
# 5) Substituindo o código 9 por NaN nas colunas ESTCIVMAE e CONSULTAS

df['ESTCIVMAE'] = df['ESTCIVMAE'].replace(9, np.nan)
df['CONSULTAS'] = df['CONSULTAS'].replace(9, np.nan)

print(df)
```

```python
# 6) Substitua os valores faltantes da quantitativa (QTDFILVIVO) por zero.

df['QTDFILVIVO'] = df['QTDFILVIVO'].fillna(0)

print(df)
```

```python
# 8) seu código aqui

df = pd.read_csv('SINASC_RO_2019.csv')

if 'APGAR5' in df.columns:
    def categorizar_apgar5(valor):
        if valor >= 8:
            return 'normal'
        elif valor >= 6:
            return 'asfixia leve'
        elif valor >= 4:
            return 'asfixia moderada'
        else:
            return 'asfixia severa'


    df['Categoria_Apgar5'] = df['APGAR5'].apply(categorizar_apgar5)
   
    frequencias = df['Categoria_Apgar5'].value_counts()

    print("\nDataFrame com categorias de Apgar:")
    print(df[['APGAR5', 'Categoria_Apgar5']])
    
    print("\nFrequências das categorias de Apgar5:")
    print(frequencias)
```
