####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados
sinasc_data = pd.read_csv('SINASC_RO_2019.csv')

# Boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(sinasc_data['IDADEMAE'])
plt.title('Boxplot da Idade da Mãe')
plt.ylabel('Idade da Mãe')
plt.xticks([1], ['Idade da Mãe'])
plt.show()
```
